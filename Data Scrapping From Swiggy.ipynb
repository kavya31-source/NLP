## Data Scrapping

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
## Data gets fetched from the given url in a text format
url="https://www.swiggy.com/bangalore/top-rated-collection"
page=requests.get(url).text
```

```python
## The text has been converted to Html format since Beautiful soup works with HTML format.
html = BeautifulSoup(page, 'html.parser')
```

```python
## Fetching Data
name=[]
rating=[]
tt=[]
price=[]
rest_name=html.find_all("div",class_="nA6kb")
rate=html.find_all("div",class_="_9uwBC wY0my")
time=html.find_all("div",class_="_3Mn31")
price_2=html.find_all("div",class_="nVWSi")
for i,j,k,l in zip(rest_name,rate,time,price_2):
    name.append(i.text.strip())
    rating.append(j.text.strip())
    tt.append(k.text.strip().split()[0].split('•')[1])
    price.append(l.text.strip().split()[0].split("₹")[1])
```

```python
## Converting the fetched data into a DataFrame
pd.DataFrame({"Restaurants":name,"Rating":rating,"Time for Delivery":tt,"Price for 2":price})
```
